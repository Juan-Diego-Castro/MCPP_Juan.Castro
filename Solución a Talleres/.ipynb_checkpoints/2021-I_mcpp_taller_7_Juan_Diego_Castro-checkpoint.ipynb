Pista: El siguiente código puede serle útil.

```python
# Create function to strip time from date field, and use it to create another column
def to_day(timestamp):
    return timestamp.replace(minute=0,hour=0, second=0)

df['Day'] = df['Date'].apply(to_day)
```

```python
df['Date'] =pd.to_datetime(df["Date"])
```

```python
df
```

```python
crimes_by_day = df.groupby('Day')
crimes_by_day['ID'].agg('count')
```

```python
crimes_by_day_community = df.groupby(['Day', 'Community Area'])
crimes_by_day_community_count = crimes_by_day_community['ID'].agg('count')
crimes_by_day_community_count
community_crime_timeseries = crimes_by_day_community_count.unstack('Community Area')
community_crime_timeseries
```

### 4.

Descargue la base de datos de información socioeconómica (https://data.cityofchicago.org/Health-Human-Services/Census-Data-Selected-socioeconomic-indicators-in-C/kn9c-c2s2).

Cree una tabla que agregue el número de crímenes por Community Area. Una esa tabla con la de datos socioeconómicos y cree un "scatter plot" de número de crímenes vs ingreso per cápita. Explique la relación en palabras.

```python
sc = pd.read_csv (r'C:\Users\MAGG\Documents\GitHub\MCPP_Juan.Castro\Solución a Talleres\Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012.csv')
sc
```

```python
sc.rename(columns={'Community Area Number': 'Community Area'}, inplace=True)
sc
```

```python
sc = sc.set_index('Community Area')
sc
```

```python
crimes_community = df.groupby(['Community Area'])
crimes_community_count = crimes_community['ID'].agg('count')
crimes_community_count

```

```python
com_gdp = sc['PERCAPITAINCOME ']
com_gdp
```

```python
newdata=pd.merge(crimes_community_count, com_gdp, how='outer', left_index=True, right_index=True)
newdata
```

```python
newdata.rename(columns={'ID': 'Crimes'}, inplace=True)
newdata
```

```python
newdata.plot.scatter(s=400, c='red', x = 'PERCAPITAINCOME ', y = 'Crimes')
```

### El ingreso per capita no explica el número de crímenes, pues a medida que aumentan los ingresos, no existe una variación significativa de los crímenes. Uno pensaría que a medida que aumenta el ingreso disminuirían los crímenes y viceversa, pero incluso la comunidad con mayor ingreso per capita, cuenta con un número de crímenes similar a comunidades con menores ingresos. Cabe resaltar la presencia de un outlier con muchos crímenes y poco nivel de ingresos.
